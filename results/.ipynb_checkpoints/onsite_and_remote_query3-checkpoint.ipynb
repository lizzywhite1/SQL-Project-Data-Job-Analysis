```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_onsite = pd.read_csv('query_3_onsite.csv')
df_remote = pd.read_csv('query_3_remote.csv')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,7))
fig.suptitle('The 10 most demanded skills for ML jobs', fontsize = 20)
fig.supxlabel('Demand Count')
fig.supylabel('Skills')

ax = sns.barplot(
    ax = axes[0],
    data = df_onsite,
    x = 'demand_count',
    y = 'skills',
    hue = 'skills',
    palette='Blues_r'
)
ax.set(title = 'Onsite ML roles', xlabel = None, ylabel = None)

ax = sns.barplot(ax = axes[1],
            data = df_remote,
            x = 'demand_count',
            y = 'skills', 
            hue = 'skills',
            palette='Blues_r'
)
ax.set(title = 'Remote ML roles', xlabel = None, ylabel = None)

fig.tight_layout()
fig.savefig('query_3_fig.png')
fig.show()
```
