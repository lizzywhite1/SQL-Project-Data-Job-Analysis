```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

## Query 2

```python
df = pd.read_csv('results_csv/query_2_top_paying_skills.csv')
```

```python
df
```

```python
sns.barplot(
    df,
    x='skill_count',
    y='skills',
    hue = 'skills',
    legend = False,
    orient = 'h',
    errorbar = None,
    palette = 'Blues_r'
)
plt.ylabel('Frequency')
plt.xlabel('Skill')
plt.title('Skill count for top 50 highest paying ML jobs in 2023', pad = 10)

plt.tight_layout()
plt.savefig('images/query_2_fig.png')
plt.show()
```

## Query 3

## Query 6

```python
df = pd.read_csv('results_csv/query_6_common_times.csv')
```

```python
df = df.sort_values('month_posted') # Sort by month 
```

```python
df.reset_index(drop = True,inplace = True)
df 
```

```python
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
```

```python
def int_to_month(numeric): 
    return months[numeric-1] # Function to assign each number to relevant month
```

```python
df['month_posted'] = df['month_posted'].apply(int_to_month)
```

```python
df
```

```python
sns.barplot(
    df,
    x='month_posted',
    y='jobs_posted',
)

plt.ylabel('Job count')
plt.xlabel(None)
plt.tick_params(axis = 'x',rotation = 45)
plt.title('Number of ML job postings per month in 2023', pad = 10)

plt.tight_layout()
plt.savefig('images/query_6_fig.png')
plt.show()
```

## Query 7

```python
df = pd.read_csv('results_csv/query_7_optimal_skills.csv')
```

```python
df
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,7))
fig.suptitle('The most optimal ML skills based on demand count and average salary', fontsize = 20)
fig.supylabel('Skills')

ax = sns.barplot(
    ax = axes[0],
    data = df,
    x = 'demand_count',
    y = 'skills',
    hue = 'skills',
    orient = 'h',
    palette='Blues_r'
)
ax.set(title = 'Demand count of most optimal skills', xlabel = 'Demand Count', ylabel = None)

ax = sns.barplot(ax = axes[1],
            data = df,
            x = 'avg_salary',
            y = 'skills', 
            hue = 'skills',
            palette='Blues_r'
)
ax.set(title = 'Average Salary of most optimal skills', xlabel = 'Average Salary ($USD)', ylabel = None)
ax.set_xlim(120000,155000)

fig.tight_layout()
fig.savefig('images/query_7_fig.png')
```
